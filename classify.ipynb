```python
import pandas as pd
import numpy as np

# Load the CSV file
file_path = 'C:/Users/Dell/OneDrive/Desktop/BME PROJ 1/data1.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
print(data.head())

# List of columns to check and correct
columns_to_check = [
    'Z_mean', 'Z_median', 'Z_weighted_avg', 'Z_geometric_mean', 'Z_harmonic_mean',
    'Deg_mean', 'Deg_median', 'Deg_weighted_avg'
]

# Function to recalculate and correct statistics for each row
def recalculate_and_correct_statistics(data):
    # Assuming you have raw data columns 'Z' and 'Deg'
    if 'Z' in data.columns and 'Deg' in data.columns:
        # Recalculate values for each row
        data['Z_mean_corrected'] = data['Z']
        data['Z_median_corrected'] = data['Z']
        data['Z_weighted_avg_corrected'] = data['Z']
        data['Z_geometric_mean_corrected'] = data['Z']
        data['Z_harmonic_mean_corrected'] = data['Z']
        
        data['Deg_mean_corrected'] = data['Deg']
        data['Deg_median_corrected'] = data['Deg']
        data['Deg_weighted_avg_corrected'] = data['Deg']
        
        # Replace incorrect values with corrected values
        data['Z_mean'] = data['Z_mean_corrected']
        data['Z_median'] = data['Z_median_corrected']
        data['Z_weighted_avg'] = data['Z_weighted_avg_corrected']
        data['Z_geometric_mean'] = data['Z_geometric_mean_corrected']
        data['Z_harmonic_mean'] = data['Z_harmonic_mean_corrected']
        
        data['Deg_mean'] = data['Deg_mean_corrected']
        data['Deg_median'] = data['Deg_median_corrected']
        data['Deg_weighted_avg'] = data['Deg_weighted_avg_corrected']
        
        # Drop the corrected columns
        data.drop(columns=[
            'Z_mean_corrected', 'Z_median_corrected', 'Z_weighted_avg_corrected', 
            'Z_geometric_mean_corrected', 'Z_harmonic_mean_corrected', 
            'Deg_mean_corrected', 'Deg_median_corrected', 'Deg_weighted_avg_corrected'
        ], inplace=True)
        
        return data
    else:
        print("Raw data columns 'Z' and 'Deg' are not present in the dataset.")
        return data

# Recalculate and correct statistics
data_corrected = recalculate_and_correct_statistics(data)

# Display the corrected values
print(data_corrected.head())

# Save the corrected data to a new CSV file if needed
data_corrected.to_csv('corrected_data1.csv', index=False)
```
